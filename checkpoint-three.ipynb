## Irregular Data

Detect outliers in your dataset and handle them as needed. Use code comments to make notes about your thought process.

```python
# (1) column 'Salary Estimate'

# I already verified there is no null in the column from above. 
# I'd like to check if there is any '-1' value in the Salary column because I saw '-1' in other columns through df.head code earlier.
# I will pull any rows that contain '-1' in the Salary column.

df[df['Salary Estimate']=='-1']

# Result: found 1 row. 
# This is not a major impact among 2253 rows. Let's drop this and create a new dataframe.
```

```python
# Create a new dataframe without the row above
df1 = df[df['Salary Estimate'] !='-1']

# Verify the result
len(df1)

# Result 2252. The total number of row has been dropped to 2252 from 2253. 
```

```python
# Now I wonder how many rows are like this, especially for two other columns that I need for my analysis -- Location, Industry.
# (2) column 'Location'

df1[df1['Location']=='-1']

# Result: not found
# Good! There is no null, no -1 in the column. This column looks solid and clean!
```

```python
# (3) column 'Industry'

df1[df1['Industry']=='-1']

# Result: found 353 rows.
# This is quite a number among 2253 rows, which takes 16%. I don't want to lose all info that is listed on the same row such as Salary and Location. 
# For this case, I'm going to replace it to string 'Unknown' instead of dropping rows.
```

## Unnecessary Data

Look for the different types of unnecessary data in your dataset and address it as needed. Make sure to use code comments to illustrate your thought process.

```python
# # I'm going to drop all unnecessary columns, keep only Salary Estimate, Location, and Industry
# # Save it as a new dataframe

df2 = df1.drop(columns=['Unnamed: 0', 	'Job Title',  'Job Description', 	'Rating',
                        'Company Name', 	'Headquarters', 	'Size', 	'Founded', 	'Type of ownership', 
                        'Sector', 	'Revenue', 	'Competitors', 	'Easy Apply'])
df2
```

## Inconsistent Data

Check for inconsistent data and address any that arises. As always, use code comments to illustrate your thought process.

```python
# Although all values in my new dataframe is not really inconsistent, I'd like to make it more flat and clean.

# Where I'm going to clean is..

# 1. Remove (Glassdoor est.) in the first column
# 2. Split the first column into two columns for each number
# 3. Split the Location by comma so that I have City and State information separately.
```

```python
# 1. Remove ' (Glassdoor est.)'

df2['Salary Estimate'] = df2['Salary Estimate'].map(lambda x: x.lstrip('').rstrip(' (Glassdoorest.)'))
df2['Salary Estimate'].to_frame()
```

```python
# 2. Split Salary Estimate to two columns

df2[['Min', 'Max']] = df2['Salary Estimate'].str.split('-', expand=True)
df2[['Min', 'Max']] 
```

```python
# 3. Split the Location

df2[['City', 'State','test']] = df2['Location'].str.split(', ', expand=True)
df2[['City', 'State', 'test']] 

# I tried to split into two columns but resulted errors because there is an additional column created by spliting for some reason.
# I just went ahead, created 3 columns. I will drop the last column.
```

```python
# 3-1. Drop test column
df3 = df2.drop(columns=['test'])
df3
```
